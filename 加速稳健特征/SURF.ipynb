# 加速稳定特征

### 加速稳定特征SURF是SIFT算法的加速版本

### 在SIFT算法当中使用高斯差分方程（Difference of Gaussian）对高斯拉普拉斯方程( Laplacian of Gaussian）进行近似。

### SURF算法计算关键点的尺度和位置信息使用Hessian矩阵实现。

```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt

img = cv2.imread("lena_std.tif")
img= cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #转化成RGB通道图像

surf = cv2.xfeatures2d.SURF_create(400) #创建surf特征点检测类，参数为hessian矩阵的阈值
key_query,desc_query = surf.detectAndCompute(img,None) #找到关键点和描述符
img=cv2.drawKeypoints(img,key_query,img) #把特征点标记到图片上
plt.imshow(img)
```
