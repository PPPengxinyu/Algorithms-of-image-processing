# 图像梯度算法  laplacian算子与Sobel算子、scharr算子

```python
import cv2   
import numpy as np   
from matplotlib import pyplot as plt 
```

```python
# 加载原图
img = cv2.imread('lena_Gray.tif',cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap='gray'),plt.title('Original')
```

## 采用Lapacian算子

### Lapacian算子的卷积核 

```python
# 使用Lapacian算子进行卷积，提取图像边缘
laplacian = cv2.Laplacian(img,cv2.CV_64F)
plt.imshow(laplacian, cmap='gray'), plt.title('laplacian')
```

## 采用Sobel算子

### Sobel算子的卷积核 
### Sobel算子算法的优点是计算简单，速度快。但是由于只采用了2个方向的模板，只能检测水平和垂直方向的边缘，因此这种算法对于纹理较为复杂的图像，其边缘检测效果就不是很理想。

```python
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
plt.subplot(121), plt.imshow(sobelx, cmap='gray'), plt.title('sobelx')
plt.subplot(122), plt.imshow(sobely, cmap='gray'), plt.title('sobely')
plt.show()
```

## 采用Scharr算子
### scharr算子临近像素的权重更大，故精确度更高，能计算出更小的梯度变化。

```python
scharrx=cv2.Scharr(img,cv2.CV_64F,1,0)
scharrx=cv2.convertScaleAbs(scharrx)
scharry=cv2.Scharr(img,cv2.CV_64F,0,1)
scharry=cv2.convertScaleAbs(scharry)
plt.subplot(121), plt.imshow(scharrx, cmap='gray'), plt.title('scharrx')
plt.subplot(122), plt.imshow(scharry, cmap='gray'), plt.title('scharry')
plt.show()
```
